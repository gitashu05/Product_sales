```python
df['Gender'].replace(['F','M'],['Female','Male'],inplace=True) # replace the F and M with Female and Male
```

### EDA

```python
#Gender Count
snt=sns.countplot(data=df,x='Gender',width=0.5)
for bars in snt.containers:  # displaying the number count on each label
    snt.bar_label(bars)
```

```python
#Sales in gender_category

sales_by_gen=df.groupby(by='Gender',as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=sales_by_gen,x='Gender',y="Amount",width=0.5)
```

##### From above data, Female are shopping more than male.

```python
#Age-Group data 
sn=sns.countplot(data=df,x='Age Group',width=0.9,hue='Gender')
for bars in sn.containers:  # displaying the number count on each label
    sn.bar_label(bars)
```

```python
# Total amount VS Agegroup
sales_age_group=df.groupby(by='Age Group',as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(data=sales_age_group,x='Age Group',y="Amount",width=0.5)
plt.show()
```

**from the above graph we can see 26-35 age group did most of the shopping**

```python
#State wise Amount spent

sales_by_state=df.groupby(by=['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(17,6)})
sns.barplot(data=sales_by_state,y="State",x="Amount")
plt.show()
```

```python
#Top 5 states by Orders

sales_by_state=df.groupby(by='State',as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(17,6)})
sns.barplot(data=sales_by_state,x='State',y="Orders",width=0.7)
plt.show()
```

**from above 2 scenarios , it is clearly observed that kerla is ordered more than Haryana but Haryana spent more Amount for purchage than kerla.**

```python
df['Marital_Status'].replace([0,1],['Bachelor','Married'],inplace=True)
```

```python
#shopping(marital vs bachelor)

sales_marital_bachelor=df.groupby(by=['Gender','Marital_Status'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
# sales_marital_bachelor
sns.set(rc={'figure.figsize':(6,4)})
sns.barplot(data=sales_marital_bachelor,x='Marital_Status',y='Orders',hue='Gender')
plt.show()
```

**most of the shopping done by bachelors(female > men) than Married(female > male)** 

```python
# Occupation wise Purchase
sns.set(rc={'figure.figsize':(25,6)})
# print(df['Occupation'].head(10))
occ_pur=sns.countplot(data=df,x='Occupation',width=0.85)
for bars in occ_pur.containers:
    occ_pur.bar_label(bars)
```

**IT sector has more purchasing Power than any other sector**

```python
# Product_Category wise Ordered

sns.set(rc={'figure.figsize':(25,8)})
sales_category=sns.countplot(data=df,x='Product_Category',width=0.9)
for bars in sales_category.containers:
    sales_category.bar_label(bars)
```

```python
# Product_Category wise sales 

Sales_Category=df.groupby(by=['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,6)})
sns.barplot(data=Sales_Category,x='Product_Category',y='Amount')
```

**From above two , it is observed that Amount invested (Food > Cloth) but Orders (cloth > food)**

```python
#Top Selling Product

Top_selling=df.groupby(by=['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.barplot(data=Top_selling,y='Product_ID',x='Orders')
```

#### Insights
- **bachelors from UP,MH,KA with age group of 26-35 working in IT sector, Health Care and Aviation shopping more on Food,Cloths,Electronics.**
